1. Create a NumPy array of shape (5, 5) with random integers between 1 and 50.

o Find the row with the maximum sum.

o Replace all elements greater than the mean with 0

```python
# importing NumPy
import numpy as np

# creating 5*5 array

random_arr= np.random.randint(1,51, size=(5,5))  # using random variables from 1 to 50
print(" Generated 5*5 random array : \n", random_arr)
```

```python
# Find row with max sum
sum_rows = random_arr.sum(axis=1)  # gives sum of each row
print( "Sum of each row:", sum_rows)                    # display each sum

max_index= np.argmax(sum_rows)  # gives the index of the row with max sums
print("index of row with max sum:", max_index)

row_max= random_arr[max_index]   #gives the row with max sum
print("row with max sum:", row_max)
```

```python
# Repalce all elements greater than mean with 0

mean_arr= random_arr.mean()  #calculates the mean
print("the mean of generated array:", mean_arr)

new_arr= random_arr.copy()  # creates a copy initial array
new_arr[new_arr > mean_arr]=0 #updating all elements > mean to 0
print("Updated array: \n",new_arr)

```

3. Load the dataset Iris data
o Display the first 10 rows and basic statistics.

o Filter all rows where sepal_length > median and species is "versicolor".

o Group the data by species and compute the mean petal_length.


```python
# importing libraries
import seaborn as sns
import pandas as pd

#load dataset Iris from seaborn
data= sns.load_dataset("iris")
print(data)
```

```python
# display first 10 rows and basic stats
first_ten= data.head(10) # gets first ten rows
print("First 10 rows: \n", first_ten)

print("\n")

#dispaly basic stats
basic_stats=data.describe()
print("Basic Statistics: \n", basic_stats)
```

```python
# Filter all rows with sepal_length > median and versicolor

median_sl=data["sepal_length"].median()   # gives median of sepal length
print(" median of sepal lenght:", median_sl)

new_data= data[(data["sepal_length"] > median_sl) & (data["species"] == "versicolor")] #selecting rows above median and species versicolor
print( "The new filtered data is:\n", new_data)
```

Add a new column named petal_ratio = petal_length / petal_width.

o Find the top 5 rows with the highest petal_ratio

```python
#Adding new column

data["petal_ratio"] = data["petal_length"]/data["petal_width"]
print("Updated data is: \n", data)

print("\n")

# Finding highest petal ratio
petal_ratio_order= data.sort_values(by="petal_ratio", ascending=False) # sorting the petal ratio from largest to smallest
highest_petal_ratio=petal_ratio_order.head(5)   # taking the largest 5
print(" The top 5 rows with highest petal ratio is: \n" ,highest_petal_ratio)
```

Using Matplotlib, create:
o A histogram of sepal_length with 15 bins.
o A scatter plot of sepal_length vs petal_length, color-coded by species

```python
# importing libraries
import matplotlib.pyplot as plt

# Creating Histogram
plt.hist(data["sepal_length"], bins=15, color="green") # setting bins as 15 and colour of graph as green

# add title and label
plt.xlabel("Sepal length")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal length with 15 bins")
plt.show()
```

```python
# Scatter of sepal vs petal length , color coded by species

# Plotting scatter for setosa
plt.scatter(data[data["species"]=="setosa"]["sepal_length"],data[data["species"]=="setosa"]["petal_length"], color="cyan", label="setosa")

#plotting scatter for versicolor
plt.scatter(data[data["species"]=="versicolor"]["sepal_length"],data[data["species"]=="versicolor"]["petal_length"], color="purple", label="versicolor")

#plotting scatter for virginica
plt.scatter(data[data["species"]=="virginica"]["sepal_length"],data[data["species"]=="virginica"]["petal_length"], color="blue", label="virginica")


# add title and label
plt.xlabel("Sepal length")
plt.ylabel("Petal length")
plt.title("Sepal length vs Petal length, colour coded by species")
plt.legend()
plt.show()
```

Using Seaborn, plot:
o A pairplot of the entire dataset.
o A boxplot showing petal_width distribution for each species.


```python
# lib are already imported above

# plotting pairplot for entire dataset
sns.pairplot(data, hue="species", markers="D")  # hue separates species with different colors

plt.show()
```

```python
# plotting box plot showing petal_width distribution for each species

sns.boxplot(data=data, x="species", y="petal_width", hue="species")

#add title
plt.title("Petal width distribution for each species") # hue separates species with different colors
plt.show()
```
